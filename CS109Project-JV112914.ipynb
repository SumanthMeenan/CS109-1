Ideally, we would like to use zip code to give us more abiltiy to break down the data, but most of the community health type indicators are by county.  To address, this we will need to use a crosswalk.  However, we run into cases where a zip code can span into multiple counties.  

For our purpose, the state of NY health website has a dataset where they mapped zip codes to county using a logic where they looked by zipcode and the assigned it to the county that contained the largest portion of the zip code.  This seems like a reasonable assumption used by the state website and since we are using datasets from the same website, we will use it here.


```python
# we will use some similar code from the 9-23 data scraping lecture

# first read in the HTML List to get all of the state names
url = 'http://www.health.ny.gov/statistics/chac/perinatal/county_list.htm'
source = urllib2.urlopen(url).read()

## convert this to a beautiful soup bs4 object
NYcounties = bs4.BeautifulSoup(source)

## get all links in the page
link_list = [l.get('href') for l in NYcounties.findAll('a')]

# create an empty list to collect the valid links
external_links = []

# write a loop to filter the links
# we want the links that start with '/statistics/chac/perinatal/county'
for l in link_list:
    if l is not None and l[:33] == '/statistics/chac/perinatal/county':
        external_links.append(l)

# now have a list of all links - we will read each one in and create a panda dataframe with county name and zipcode

# first create a blank dataframe to store results
countyzip = []
columns = ['County','Zip']
countyzip = pd.DataFrame(columns = columns)

# go through the linklist
j = 0
iterations = len(external_links)
while j < (iterations):
    # read in url name
    html = 'http://www.health.ny.gov'
    name = html + external_links[j]
    countyname = urllib2.urlopen(name).read()
    # Get county name
    # unfortunately, the HTML tag for county has the string 'county' in it.   we just want the name.   
    # Will parse based on url name - always found in same location

    # County name will be between the string 'county/' and '.htm' in the http string
    ## find "county/"
    positionstart =  name.find('county/')

    ## find ".htm"
    positionend =  name.find('.htm')

    # Get County name.  the start string 'county/' is 7 characters long - factor in
    county = name[positionstart+7:positionend]



    ## convert this to a beautiful soup bs4 object
    Countyinfo = bs4.BeautifulSoup(countyname)


    # loop over the table rows (tr tag) and get the text of cells (td tag) inside   zip is the first column.  
    # need to remove first two table row names - they are headers in the table
    for tr in Countyinfo.find_all('tr')[2:]:
        tds = tr.find_all('td')
        Zipres = tds[0].text
        Countyres = county.capitalize()
        countyzip=countyzip.append({"County": Countyres, "Zip":Zipres}, ignore_index=True)
        # last row is the total - return the dataframe without the total for cleanliness
        #countyzip=countyzip.drop(countyzip.tail(1).index)

    j = j + 1

# convert zipcode to an integer and drop the total rows
countyzip = countyzip[countyzip.Zip != 'Total']
countyzip = countyzip[countyzip.Zip != 'Total ']
countyzip['Zip'] = countyzip['Zip'].astype(int)

# note, there are two counties that have a space in here - need to adjust the string to allow proper matching
countyzip.County[countyzip.County=='Newyork'] = 'New York'  # format New York county
countyzip.County[countyzip.County=='Stlawrence'] = 'St. Lawrence'  # format St. Lawrence county
```

Need to do some integreity checks against the results.  we need to check to see if there are zip codes not matched in the ED utilization file and then add on the county code

```python
# check to see if there are zip codes not in our mapping file
# do this by creating a subset of the ED use, compare to this countyzip file, and see what is left
# just check one year (2012) for a complete list
EDfilezipchk = EDusejoined[['discharge_year','patient_zipcode','2010 Census Population']]
EDfilezipchk = EDfilezipchk[EDfilezipchk.discharge_year != 2011]

# merge this with a left join
zipcheck = pd.merge(EDfilezipchk, countyzip, how="left", left_on="patient_zipcode", right_on="Zip")

# do a quick groupby to check those that match and those that don't
# provide total census of all
All_pop = zipcheck['2010 Census Population'].sum()
# print those we do not have a match for "nan" in the Zip file merged
Missing = zipcheck[zipcheck.Zip.isnull()]
Missing_Pop = Missing['2010 Census Population'].sum()

print All_pop
print Missing_Pop
print Missing
```

```python
# merge the ED indicators and append on the county code - drop those without a match
# merge this with a left join
EDuseWithCounty = pd.merge(EDusejoined, countyzip, how="inner", left_on="patient_zipcode", right_on="Zip")

# drop the duplicate zipcode from the census data on the merge
EDuseWithCounty.drop('Zip', axis=1, inplace=True)
```


Now lets read in some of the community health variables (which are by county) and try to marry them together at the county level

```python
### JUST TO DO EXPLORATORY - WILL READ IN THE 2008-2012 NON API RESULTS
# note this is by county - so need to match to that result

# first provide the URL string with the json location
url_str = "http://health.data.ny.gov/resource/tchg-ruva.json"

# then read json file as a pandas dataframe
CommDiabInd= pd.read_json(url_str)

# county codes 1-62 are true counties - subset data
CommDiabInd = CommDiabInd[CommDiabInd.county_code < 100]

#  merge to the ED target dataset - call it DiabIndicators for now
DiabIndicators = pd.merge(EDuseWithCounty, CommDiabInd, how="inner", left_on="County", right_on="county_name")
```

Now that we have some indicators, let's do some basic exploratory analysis to start.  Our target is looking at overutilizers of ED, which for our purposes will be the "difference_in_rates" column.   Let's get an idea of the distribution:


```python
# plot the distribution of the rates across zip codes in the state - do 
df = EDuseWithCounty[EDuseWithCounty['discharge_year'] == 2012]
plt.plot(subplots=True)
plt.hist(df.difference_in_rates.values, bins = 20, color='b')
plt.title('Difference in actual to expected ED use rates for 2012')
plt.xlabel('Actual-Expected')
plt.ylabel('Frequency')
plt.show()

# plot the distribution of the rates across zip codes in the state - do 
df = EDuseWithCounty[EDuseWithCounty['discharge_year'] == 2011]
plt.plot(subplots=True)
plt.hist(df.difference_in_rates.values, bins = 20, color='r')
plt.title('Difference in actual to expected ED use rates for 2011')
plt.xlabel('Actual-Expected')
plt.ylabel('Frequency')
plt.show()
```

```python
# noticing a trend where the majority of zips are actually using the ED less than expected - lets see if we have correlations by diabetic indicator

# first, let's assign our target variable if the difference in rates is >0 (indicating a place we are trying to explore) 
EDuseWithCounty['ED Opp'] = EDuseWithCounty['difference_in_rates'].apply(lambda x: 0 if x < 0 else 1)
DiabIndicators['ED Opp'] = DiabIndicators['difference_in_rates'].apply(lambda x: 0 if x < 0 else 1)

# now lets subset our results by indicator - see what trends we see
# For now, just manually picking an indicator to get a feeling
# start with g29-Age-adjusted percentage of adults with physician diagnosed diabetes
# graph percentage rate to our target of difference in rates as a scatterplot

df = DiabIndicators[DiabIndicators['indicator_number'] == 'g29']
plt.scatter(df['difference_in_rates'], df['percentage_rate'])
plt.title('ED use by % of population diagnosed with diabetes')
plt.xlabel('ED use Actual-Expected')
plt.ylabel('% of popluation with diabetes')
plt.grid()
plt.show()

# boxplot comparing the distribution of the two
df.boxplot(column='percentage_rate', by='ED Opp', grid=False)
plt.ylabel('% of popluation with diabetes')
```
